```python
import pandas as pd

# Load the CSV with semicolons
df = pd.read_csv('Dataset Denggi Malaysia.csv', sep=';')

# Inspect the data
print("Columns:", df.columns)
print(df.head(3))

# Convert wide → long format
df_long = df.melt(id_vars=['state/year'], var_name='Year', value_name='Cases')

# Rename column
df_long.rename(columns={'state/year': 'State'}, inplace=True)

# Convert Year to int
df_long['Year'] = df_long['Year'].astype(int)

# Sort by State and Year
df_long = df_long.sort_values(['State', 'Year']).reset_index(drop=True)

# Create lag features
df_long['Cases_last_year'] = df_long.groupby('State')['Cases'].shift(1)
df_long['Cases_2yrs_ago'] = df_long.groupby('State')['Cases'].shift(2)

# 3-year rolling average
df_long['Cases_3yr_avg'] = df_long.groupby('State')['Cases'].rolling(window=3, min_periods=1).mean().reset_index(0, drop=True)

# Fill missing values for first years
df_long.fillna(0, inplace=True)

# Save ML-ready CSV
df_long.to_csv('Denggi_Malaysia_ML_Ready.csv', index=False)

# Preview the first 15 rows
print(df_long.head(15))
```

```python
import pandas as pd

# LOAD DATA

weather_df = pd.read_csv("malaysia_weather.csv")

# STANDARDIZE COLUMN NAMES

weather_df.columns = (
    weather_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print("Cleaned column names:", weather_df.columns.tolist())

# RENAME IMPORTANT COLUMNS

rename_map = {
    'state': 'State',
    'year': 'Year',
    'total_rainfall_in_millimetres': 'TotalRainfall',
    'mean_relative_humidity_in_percentage': 'Humidity'
}

weather_df = weather_df.rename(columns=rename_map)


# CREATE AvgTemp SAFELY

if 'avgtemp' not in weather_df.columns:

    min_temp_col = [c for c in weather_df.columns if 'minimum' in c and 'temperature' in c]
    max_temp_col = [c for c in weather_df.columns if 'maximum' in c and 'temperature' in c]

    if min_temp_col and max_temp_col:

        # Convert to numeric FIRST
        weather_df[min_temp_col[0]] = pd.to_numeric(weather_df[min_temp_col[0]], errors='coerce')
        weather_df[max_temp_col[0]] = pd.to_numeric(weather_df[max_temp_col[0]], errors='coerce')

        # Compute average
        weather_df['AvgTemp'] = (
            weather_df[min_temp_col[0]] + weather_df[max_temp_col[0]]
        ) / 2

    else:
        print("ERROR: Could not find min/max temperature columns.")


# KEEP ONLY REQUIRED FEATURES

weather_df = weather_df[['State', 'Year', 'AvgTemp', 'TotalRainfall', 'Humidity']]

# DROP MISSING ROWS

weather_df = weather_df.dropna()

# SAVE CLEAN FILE
weather_df.to_csv("clean_weather_dataset.csv", index=False)

print("\nDONE! Final dataset saved as clean_weather_dataset.csv")
print(weather_df.head())



```

```python
import pandas as pd

# Load raw dataset (the one you showed)
df = pd.read_csv("clean_weather_dataset.csv")

# Replace '-' with NaN
df = df.replace("-", pd.NA)

# Convert numeric columns properly
num_cols = ["AvgTemp", "TotalRainfall", "Humidity"]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Group by State-Year and average all weather values
df_clean = df.groupby(["State", "Year"], as_index=False).mean()

# Drop rows with missing values
df_clean = df_clean.dropna()

# Save final dataset
df_clean.to_csv("weather_final.csv", index=False)

df_clean.head()
```

```python
import pandas as pd

# Load datasets
dengue_df = pd.read_csv('Denggi_Malaysia_ML_Ready.csv')
weather_df = pd.read_csv('weather_final.csv')

# Normalize State names (convert both to uppercase)
dengue_df['State'] = dengue_df['State'].str.upper()
weather_df['State'] = weather_df['State'].str.upper()

# Ensure 'Year' is integer
dengue_df['Year'] = dengue_df['Year'].astype(int)
weather_df['Year'] = weather_df['Year'].astype(int)

# Filter weather dataset to match dengue dataset years (2000–2019)
weather_df = weather_df[(weather_df['Year'] >= 2000) & (weather_df['Year'] <= 2019)]

# Merge datasets on State and Year
merged_df = pd.merge(dengue_df, weather_df, on=['State', 'Year'], how='inner')

# Check the merged dataset
print(merged_df.head())

#save merged dataset
merged_df.to_csv('dengue_weather_merged.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import numpy as np

# Load CSV
df = pd.read_csv("dengue_weather_merged.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Encode 'State'
le = LabelEncoder()
df['State'] = le.fit_transform(df['State'])

# Feature scaling for numeric columns (except target 'Cases')
num_features = ['Cases_last_year','Cases_2yrs_ago','Cases_3yr_avg','AvgTemp','TotalRainfall','Humidity']
scaler = MinMaxScaler()
df[num_features] = scaler.fit_transform(df[num_features])

# log transform target to reduce skew
df['Cases_log'] = np.log1p(df['Cases'])  # log(1 + x) avoids log(0) issues

#  Check dataset
print(df.head())
```

```python
import pandas as pd

# Load dataset
pop_df = pd.read_csv('population_state.csv')

# Check columns
print(pop_df.columns)

# Convert 'date' to year
pop_df['Year'] = pd.to_datetime(pop_df['date'], errors='coerce').dt.year

# Filter years 2000-2019
pop_df = pop_df[(pop_df['Year'] >= 2000) & (pop_df['Year'] <= 2019)]

# Sum population over sex, age, ethnicity
pop_simple = pop_df.groupby(['state','Year'], as_index=False)['population'].sum()

# Check result
print(pop_simple.head())

# Save cleaned dataset
pop_simple.to_csv('population_simple.csv', index=False)
```

```python
import pandas as pd

# Load dengue dataset
dengue_df = pd.read_csv('dengue_weather_merged.csv')

# Load population dataset
pop_df = pd.read_csv('population_simple.csv')

# Check columns
print(pop_df.columns)  # Make sure we see the correct names

# Strip spaces and standardize state names
pop_df['state'] = pop_df['state'].str.strip().str.upper()

# Make sure 'Year' column is numeric
pop_df['Year'] = pd.to_numeric(pop_df['Year'], errors='coerce')

# Filter for years 2000-2019
pop_df = pop_df[(pop_df['Year'] >= 2000) & (pop_df['Year'] <= 2019)]

# Sum population over sex, age, ethnicity to get total population per state-year
pop_simple = pop_df.groupby(['state','Year'], as_index=False)['population'].sum()

# Standardize dengue dataset state names
dengue_df['State'] = dengue_df['State'].str.strip().str.upper()
dengue_df['Year'] = pd.to_numeric(dengue_df['Year'], errors='coerce')

# Merge population into dengue dataset
merged_df = dengue_df.merge(pop_simple, left_on=['State','Year'], right_on=['state','Year'], how='left')

# Drop duplicate 'state' column from pop dataset
merged_df = merged_df.drop(columns=['state'])

# Check for missing population values
missing = merged_df[merged_df['population'].isna()]
print("Missing population rows after merge:")
print(missing[['State','Year']].drop_duplicates())

# Save merged dataset
merged_df.to_csv('dengue_with_population.csv', index=False)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dengue_with_population.csv")

# Columns to normalize
cols_to_normalize = ['Cases', 'Cases_last_year', 'Cases_2yrs_ago', 'Cases_3yr_avg']

# Normalize each column per 100,000 population
for col in cols_to_normalize:
    df[f'{col}_per_100k'] = (df[col] / df['population']) * 100000

# Check the result
print(df.head())

# Save to a new CSV
df.to_csv("dengue_data_normalized_all.csv", index=False)
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv("dengue_data_normalized_all.csv")

# Create Outbreak column: 1 if current cases > 3-year average, else 0
df['Outbreak'] = df.apply(lambda row: 1 if row['Cases'] > row['Cases_3yr_avg'] else 0, axis=1)

# Optional: save the updated dataset
df.to_csv("dengue_dataset_with_outbreak.csv", index=False)

print(df[['State', 'Year', 'Cases', 'Cases_3yr_avg', 'Outbreak']].head(20))
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('dengue_dataset_with_outbreak.csv')

# Define lagged columns
lagged_cols = ['Cases_last_year', 'Cases_2yrs_ago', 'Cases_3yr_avg']

# Function to remove rows where lagged features are 0 or missing
def clean_state_data(state_df):
    # Remove rows where any lagged column is 0
    for col in lagged_cols:
        state_df = state_df[state_df[col] != 0]
    return state_df

# Apply per state
df_cleaned = df.groupby('State', group_keys=False).apply(clean_state_data)

# Reset index
df_cleaned = df_cleaned.reset_index(drop=True)

# Save cleaned dataset
df_cleaned.to_csv('dengue_dataset_cleaneddd.csv', index=False)

print("Cleaned dataset saved as 'dengue_dataset_cleaned.csv'")
```
